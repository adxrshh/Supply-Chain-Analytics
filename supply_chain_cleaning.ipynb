```python
# Dropping completely null columns
df.drop(columns=['Order Zipcode' , 'Product Description'] , inplace=True)
df
```

```python
# Customer Lname is irrelevant and has 8 nulls so deleting it
df.drop('Customer Lname', axis=1, inplace=True)
```

```python
df.isna().sum()
```

```python
# Deleting the 3 null rows in Customer zipcode
df = df.dropna(subset=["Customer Zipcode"])
```

```python
# NO null values now
df.isna().sum()
```

```python
# Fixing the date format
df.loc[:, "order date (DateOrders)"] = pd.to_datetime(df["order date (DateOrders)"])
df.loc[:, "shipping date (DateOrders)"] = pd.to_datetime(df["shipping date (DateOrders)"])
```

```python
df.columns
```

```python
# Now splitting our main file into 4 csv files
customers = df[["Customer Id", "Customer City", "Customer State", "Customer Country",
                "Customer Zipcode", "Customer Segment"]].drop_duplicates(subset=["Customer Id"])

products = df[["Product Card Id", "Product Name", "Product Price", "Product Status",
               "Category Id", "Category Name", "Department Id", "Department Name"]].drop_duplicates(subset=["Product Card Id"])

orders = df[["Order Id", "Order Customer Id", "order date (DateOrders)",
             "shipping date (DateOrders)", "Days for shipping (real)",
             "Days for shipment (scheduled)", "Delivery Status", "Late_delivery_risk",
             "Shipping Mode", "Order Status", "Market", "Order Region", "Order State",
             "Order City", "Order Country", "Latitude", "Longitude"]].drop_duplicates(subset=["Order Id"])

order_items = df[["Order Item Id", "Order Id", "Product Card Id", "Order Item Quantity",
                   "Order Item Discount", "Order Item Discount Rate", "Order Item Product Price",
                   "Order Item Profit Ratio", "Sales", "Order Item Total", "Order Profit Per Order",
                   "Benefit per order", "Sales per customer"]]

customers.to_csv("customers.csv", index=False)
products.to_csv("products.csv", index=False)
orders.to_csv("orders.csv", index=False)
order_items.to_csv("order_items.csv", index=False)

print(len(customers), len(products), len(orders), len(order_items))
```
